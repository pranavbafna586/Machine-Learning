```python
import numpy as np

# Dot product of two vectors
# a.b = ab cos O
a = np.array([2, 3])
b = np.array([4, 4])
a_dot_b = np.dot(a, b)
print("a_dot_b:", a_dot_b)

# Cross Product
# a x b = ab sin O n
a_cross_b = np.cross(a, b)
print("a_cross_b:", a_cross_b)

# Projection of vector a onto vector v
# projV a = (a . v / |v|^2) v
a = np.array([2, 5])
v = np.array([8, -6])

# Magnitude of vector v
mag_of_v = np.sqrt(np.sum(v**2))

# Projection of vector a onto vector v
proj_of_a_on_v = (np.dot(a, v) / mag_of_v**2) * v
print("proj_of_a_on_v:", proj_of_a_on_v)
```

# Statistics in ML

```python
#  collecting analysing ,interpretin presenting data
# 1)range
# 2)mean
# 3) Standard deviation
```

```python
# Basics of statistics and types of data
                # |
            # help to extract information from data
# Types of data 
# categoreal          Numerical
#     |                  | 
# Nominal Ordinal     Discrete continuous
```

```python
# Main types of statistics are
# Descriptive statistics--> describe the data 
#                       --> Data Analysis
#                       --> Measure of Central tendencies(Mean,Median,Mode)    
#                       --> Measure of Variability(Range,Std Deviation,Variance)    
# Inferential Statistics-->Get inferences about the data
#                       -->Data Science
```

```python
# Sampling Techniques
# 1)Simple Random Sampling --> Randomly picked from a larger population 
#                          --> all datapoints has equal probability to be selected as sample data
#                          --> Eg:- Employee survey in a company
#                          --> Population size should be high,,,No sample bias
#  
# 2)Systematic Sampling    --> Sample picked from sample at regular intervals
#                          --> carried in Homogenous and data points are uniformly distribured
#                          --> Eg:- selecting every 10 th member from population
#                          --> Quick and easy ,,less bias,,Data manipulation risk,,requires randomness in data 
# 
# 3)Stratified R Sampling  --> Population is subdivided into smaller group called Strata
#                          --> Eg:- Smartphone saled in all states
#                          --> cannot be perfomed on population classified in grps
#                          --> overlapping data points
#  
# 4)Cluster Sampling       --> Carried on population that are inherent groups
#                          --> population divided into clusters 
#                          --> Eg:- Smartphone sales in randomly selected states
#                          -->  overlapping data points,,cannot get general insight for entire population 
```

```python
# Central Tendency -->center point of dataset ,, it summarizes the data
# 1)Mean   --> avg value
#              missing values replaced by mean if data is uniform distribution
# 2)Median --> middle value (ascending order)
#              missing values replaced by median if data is skewed distribution
# 3)Mode   --> most frequent value
#              missing values replaced by Mode if data has missing categorical values
```

```python
# Measure of variability
# 1)Range --> set of data diff between largest and smallest
# 2) Variance --> average of squared diff from mean
# 3) Standard Deviation --> square root of variance
```

```python
# Purpose of these Measures
# 1)percentile
# 2)Quantile divided dataest in equal-sized subgroups
```

```python
# Correlation and causation
# extent to which two variables are related to each other in dataset
# 1)positive corelation
# Eg: height and weight of a person
# 2)negative corelation
# Eg: height and age of a person
```

```python
# Hypothesis testing 
# assumption that is made based on data 
# Null hypothesis:  = 0  accepted as fact
#              Eg: Sun rotate around planet
# Alternative hypothesis:   0 challenges the null hypothesis
#              Eg: Planet rotate around sun
```
