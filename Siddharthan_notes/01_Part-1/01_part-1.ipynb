# AI vs ML vs DL
- Ai is brader field ML is subset of Ai and DL is subset of ML
1) AI is conncerned with building smart & intelligent machines
2) ML is technique to implement AI that learn from the data by themselves being explitly programmed
3) DL subfield of ML uses ANN to learn from data

# Supervised Machine Learning
- Algo learns from labelled data
1) Classification
- predicting a class or discrete values
    Eg:- Male or Female
    - Types
    1) Decision Tree Classification 
    2) Random forest Classification
    3) k-nearest neighbour
2) Regression
- prediciting a quantity or continuous values
    Eg:- Salary ,age
    - Types
    1) Logistic Regression 
    2) Polynomial Regression
    3) Support Vector Machines

# Unsupervised Machine Learning
- Algo lears from unlabelled data
1) Clustering
- grouping similar data points
    Eg:- more phone less wifi company offer
2) Association
- used to find imp relation between data points
    Eg:- movie suggest
- Types
1) K-means Clustering
2) Hierarchical Clustering
3) Principal Component Analysis(reduce dimentions of data)
4) Apriori
5) Eclat
